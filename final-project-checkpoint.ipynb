import sklearn and necessary libraries

```python
from sklearn import preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor  
import numpy as np
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
%matplotlib inline
```

Load in our datasets, no label for the TEST

```python
XTRAIN = np.loadtxt("X_train",delimiter=",")
XTEST = np.loadtxt("X_test",delimiter=",")
y_train = np.loadtxt("y_train",delimiter=",")
```

```python
print(XTRAIN.shape)
print(y_train.shape)
print(XTEST.shape)
#shift X over 
```

Now, because we were not given labels for the test set, we are splitting the Training set 80-20. 20% allocated for testing after fitting it with the first 80%

```python
X_train, X_test, y_train, y_test = train_test_split(
...     XTRAIN, y_train, test_size=0.2, random_state=42)
```

Splice NP array in order to have a smaller dataset for easier testing (I dont have to wait all day for it to fit)

```python
shiftedxtrain = X_train[0:2000,:]
shiftedytrain = y_train[0:2000]
print(shiftedxtrain.shape)
print(shiftedytrain.shape)
```

fitting model with decision tree regressor

```python
regressor = DecisionTreeRegressor(random_state = 0)  
regressor.fit(X_train, y_train) 
predictions = regressor.predict(X_test)
```

Plot y_test vs predictions done on our "test set"

```python
plt.scatter(y_test, predictions)
plt.xlabel("True Values")
plt.ylabel("Predictions")
plt.show()
print(regressor.score(X_test, y_test))
```
